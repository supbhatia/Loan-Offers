```python
uq = pd.read_csv('Downloads/uq.csv')
```

```python
uq
```

```python
ip = 'Chinese'
cat =''
for i in range(len(uq)):
    if ip == uq.iloc[i,0]:
        cat = uq.iloc[i,1]
print(cat)
```

```python
colE = ''
if cat == 'white':
    colE = 'White_E'
elif cat == 'asian':
    colE = 'Asian_E'
elif cat == 'Hispanic':
    colE = 'Hispanic_E'
elif cat == 'African':
    colE = 'African_E'
else:
    colE = 'American_Indian_E'
print(colE)
```

```python
pop = cuisine[cuisine[colE] != 0]
```

```python
colsc = ''
if cat == 'white':
    colsc = 'white_sc'
elif cat == 'asian':
    colsc = 'Asian_sc'
elif cat == 'Hispanic':
    colsc = 'Hispanic_E'
elif cat == 'African':
    colsc = 'African_sc'
else:
    colsc = 'American_sc'
print(colsc)
```

```python
pop = pop.sort_values(by = colsc, ascending = True)
pop
```

```python
rest = pd.read_csv('Downloads/cuisine.csv')
rest
```

```python
rest_final = rest[['Row Labels',ip]]
```

```python
rest_final = rest_final.sort_values(by = ip , ascending = True)
rest_final
```

```python
result = pd.merge(pop, rest_final, how='inner', on=['Row Labels'])
```

```python
colc = ''
if cat == 'white':
    colc = 'white_C'
elif cat == 'asian':
    colc = 'asian_C'
elif cat == 'Hispanic':
    colc = 'Hispanic_C'
elif cat == 'African':
    colc = 'African_C'
else:
    colc = 'american_indian_C'
print(colc)
```

```python
result = result[result[colc] <= 10].nlargest(50,[colE])
result
```
